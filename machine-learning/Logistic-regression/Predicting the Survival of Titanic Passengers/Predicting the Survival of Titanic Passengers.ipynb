```python
def Fill_Age(data):
    age=data[0]
    sex=data[1]
    if pd.isnull(age):
        if sex is 'male':
            return 25
        else:
            return 27
    else:
        return age
```

```python
trng_set['Age']=trng_set[['Age','Sex'] ].apply(Fill_Age,axis=1)
```

```python
trng_set.drop(['Name','Ticket'],axis=1,inplace=True)
```

```python
male=pd.get_dummies(trng_set['Sex'],drop_first=True)
```

```python
trng_set.drop(['Sex'],axis=1,inplace=True)
```

```python
male.head(10)
```

```python
trng_set=pd.concat([trng_set,male],axis=1)
```

```python
X=trng_set.drop(['Survived','Embarked','Fare'],axis=1).values
```

```python
trng_set
```

```python
X[1]
```

```python
Y=trng_set['Survived'].values
```

# STEP #4: MODEL TRAINING

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.4,random_state=0 )
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(random_state=0)
classifier.fit(X_train,Y_train)
```

```python
pd.DataFrame()
```

# STEP #5: MODEL TESTING

```python
Y_predict=classifier.predict(X_test)
```

```python
Y_predict
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,Y_predict)
sns.heatmap(cm,annot=True,fmt='d')
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,Y_predict))
```
